```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from imgtools import *
```

```python
my_path = 'D:/python/AI/img/01/01.JPG'
img = Image.open(my_path)

equalized_img = histogram_equalization(img)

plt.figure(figsize=(15,10))
# Tao luoi 2x2
plt.subplot(2, 2, 1)
plt.imshow(img.convert("L"), cmap="gray")
plt.title("Anh xam goc")

plt.subplot(2, 2, 2)
plt.hist(np.array(img.convert("L")).flatten(),bins=128)
plt.title("Bieu do histogram cua anh goc")

plt.subplot(2, 2, 3)
plt.imshow(equalized_img,cmap="gray")
plt.title("Anh da can bang")

plt.subplot(2, 2, 4)
plt.hist(np.array(equalized_img).flatten(),bins=128)
plt.title("Bieu do histogram cua anh da can bang")

plt.show()
```
