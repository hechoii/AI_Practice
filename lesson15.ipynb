```python
from scipy import linalg
import numpy as np
```

```python
# Định nghĩa hệ phưuong trình tuyến tính:
A = np.array([[1,2],[5,7]])
b= np.array([12,3])

# Giải phưuogn trình tuyến tính:
x = linalg.solve(A,b)
print("Kết quả: ",x)
```

```python
# tính tích phân của hàm số:
from scipy import integrate

# Định nghĩa hàm fx = x^2
def my_function(x):
    return x**2

# tính tích phân fx từ 0 đến 1
integral = integrate.quad(my_function,0,1)
print("Tich phan cuar fx: ",integral)
```

```python
# Tính giá trị riêng và vecto riêng của ma trận
A = np.array([[2,1],
             [3,7]])

# tính giá trị riêng và vectow riêng của ma trận:
evals, evectos = linalg.eig(A)

print("Gia tri rieng: ", evals)
print("Vecto rieng: ", evectos)
```
