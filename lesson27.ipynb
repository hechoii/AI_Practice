```python
import cv2
import os
import numpy as np
import dlib
from imutils import face_utils
```

```python
path = "./img/01/face1.png"
```

```python
img = cv2.imread(path)
cv2.imshow("face",img)
cv2.waitKey(0)
cv2.destroyWindow("face")
```

```python
face_landmarks_model_path = "./moudules/shape_predictor_68_face_landmarks.dat" # data
face_recognition_model_path = "./moudules/dlib_face_recognition_resnet_model_v1.dat" # module
```

```python
# Tạo bộ nhận diện khuôn mặt
face_detector = dlib.get_frontal_face_detector()

# Đọc hình ảnh và nhận diện khuôn mặt:
im = cv2.imread(path)

# Nhận diện các khuôn mặt:
faces = face_detector(img)

# Vòng lặp qua từng khuôn mặt:
for face in faces:
    point_a = face.left(), face.top()
    point_b = face.right(), face.bottom()

    # Phát hiện ra các đặc trưng trên khuông mặt:
    predictor = dlib.shape_predictor(face_landmarks_model_path)
    face_shape = predictor(img, face)
    face_shape_array = face_utils.shape_to_np(face_shape)

    # Tính toán:
    shape_descriptor = dlib.face_recognition_model_v1(face_recognition_model_path)
    face_descriptors = shape_descriptor.compute_face_descriptor(img,face_shape)
    #print(face_descriptors)
    
    # Vẽ hcn cho khuôn mặt:
    cv2.rectangle(img,point_a,point_b,(255,255,255),1)

    # vẽ các đặc điểm khác
    for point in face_shape_array:
        cv2.circle(img,tuple(point),3,(255,0,255),-1)

cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
