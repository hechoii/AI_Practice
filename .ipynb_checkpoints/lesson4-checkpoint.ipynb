```python
from PIL import Image
from imgtools import *
from IPython.display import display
```

```python
# Đường dẫn ảnh
my_path = 'D:/python/AI/img/01/01.JPG'
```

```python
img = load_image(my_path)
display(img)
```

```python
print(img.size)
```

```python
new_img = img.resize((550,366))
display(new_img)
```

```python
# Định vị phần cắt:
region = (110, 70, 428, 323)
# Cắt vùng ảnh
crop_img = new_img.crop(region)
display(crop_img)
```

```python
# Dán
# Xác định vị trí cần dán
paste_position = (200,10)
new_img.paste(crop_img,paste_position)
display(new_img)
```
