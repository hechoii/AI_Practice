```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
from imgtools import *
```

```python
x = [1,2,3,4,5]
y = [10,16,20,30,25]

x1 = [11,8,9,7]
y1 = [15,20,25,30]
```

```python
plt.plot(x,y,"bo-",label="Dữ liệu")
plt.plot(x1,y1,"ro-",label="231")
plt.legend()
plt.xlabel("Truc x")
plt.ylabel("Truc y")
plt.show()
```

```python
# Vẽ biểu đồ cột:
categories = ["A","B","C","D","E"]
values = [12,45,32,19,66]

plt.bar(categories,values,color="c",alpha=0.6)
plt.xlabel("truc x")
plt.ylabel("truc y")
plt.show()
```

```python
# Vẽ biểu đồ tròn:
labels1 = ["A","B","C","D","E"]
size = [10,20,30,30,10]
color = ["red","green","blue","cyan","yellow"]

plt.pie(size,labels=labels1,colors=color,autopct="%1.1f%%")
plt.title("Bieu do")

plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Python is a amazing language for data visualization and word clouds"
wordcloud = WordCloud(width=800,height=400).generate(text)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Bieu do dam may")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Danh mục và giá trị
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 4, 3, 7, 6]

N = len(categories)
values += values[:1]

# Góc cho các trục
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles,values,"o-",linewidth=2)
plt.fill(angles,values,"b",alpha=0.1)

plt.title('Biểu đồ radar ví dụ')
plt.show()
```
