```python
import cv2
from imgtools import *
import time
```

```python
def detect(img):
    # Khởi tạo Haar Cascade Classifier nhận diện khuôn mặt
    face_cascade = cv2.CascadeClassifier("./data/haarcascade_frontalface_default.xml")
    eye_glass_cascade = cv2.CascadeClassifier("./data/haarcascade_eye_tree_eyeglasses.xml")

    # Chuyển sang ảnh xám
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Nhận diện mặt
    faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5, minSize=(30,30))
    # Nhận diện mắt
    eyes = eye_glass_cascade.detectMultiScale(gray_img,scaleFactor=1.1, minNeighbors=2, minSize=(5,5))

    # Vẽ hộp chứa khuôn mặt:
    for (x, y, w, h) in faces:
        cv2.rectangle(img,(x,y), (x+w,y+h), (0,255,0), 2)
    # Vẽ hình tròn chứa mắt:
    for (x, y, w, h) in eyes:
        cv2.circle(img,(int(x+ w/2), int(y+ h/2)), 20, (0,255,255), 2)

    # Trả về ảnh
    return img
```

```python
# Đọc camera
camera = camera = cv2.VideoCapture(0)
```

```python
camera = cv2.VideoCapture(0)
# Tạo cửa sổ để hiện thị
cv2.namedWindow("Video Player", cv2.WINDOW_NORMAL)
# Sau 200 khung hình lưu ảnh 1l
interval = 200
count = 0
```

```python
# Ghi text:
font = cv2.FONT_HERSHEY_SIMPLEX
font_color = (255,255,255)
font_scale = 1
font_thickness = 2
```

```python
# Hiển thị từng khung ảnh 1:
while True:
    #Thời gian trc khi đọc
    start_time = time.time()
    
    ref, frame = camera.read()
    # Thoát khi không thể đọc frame
    if not ref:
        break
        
    # Tăng count lên 1
    count+=1
    
    # Lưu ảnh sau 200fream
    if(count%interval==0):
        cv2.imwrite(f"./data/img_{count}.JPG",frame)
    
    # Lấy thời gian sau khi đọc xong
    end_time = time.time()

    # Nhận diện khuôn mặt
    detect(frame)
    
    # Tính fps
    fps = 1 / (end_time - start_time)
    
    # Ghi số lượng FPS
    cv2.putText(frame,f"FPS: {fps:.2f}",(10,200),font,font_scale,font_color,font_thickness)

    # Hiển thị khung ảnh
    cv2.imshow("Video Player",frame)
    if(cv2.waitKey(10)==ord("q")):
        break


# Huỷ bỏ player
camera.release()
cv2.destroyAllWindows()
```
