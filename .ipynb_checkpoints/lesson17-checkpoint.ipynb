```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
my_path = 'D:/python/AI/img/01/thing.JPG'
img = Image.open(my_path)
```

```python
gray_img = img.convert("L")
arr_img = np.array(gray_img)
```

```python
# Tnh gradient theo hng x:
grad_x = ndimage.sobel(arr_img, axis=0,mode="constant") # 0 l o hm theo trc x, 1 l o hm theo trc y 
grad_y = ndimage.sobel(arr_img, axis=1,mode="constant")

# Hin th nh
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(gray_img,cmap="gray")
plt.title("Anh goc")

plt.subplot(2,2,2)
plt.imshow(grad_x,cmap="gray")
plt.title("Anh lam mo theo sobel dao ham theo x") # Thng tm ra ng bin ca nh theo chiu ngang

plt.subplot(2,2,3)
plt.imshow(grad_y,cmap="gray")
plt.title("Anh lam mo theo sobel dao ham theo y") # Thng tm ra ng bin ca nh theo chiu dc
```

```python
grad_x = ndimage.convolve(arr_img, np.array([[-1,0,1],[-1,0,1],[-1,0,1]])) # Tnh gradient theo hng x 
grad_y = ndimage.convolve(arr_img, np.array([[-1,-1,-1],[0,0,0],[1,1,1]])) # Tnh gradient theo hng y

# Tnh bin  gradient:
grad_magnitude = np.sqrt(grad_x**2 + grad_y**2)

# Hin th nh
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_img,cmap="gray")
plt.title("Anh goc")

plt.subplot(2,1,2)
plt.imshow(grad_magnitude,cmap="gray")
plt.title("Anh lam mo theo Prewitt")
```
