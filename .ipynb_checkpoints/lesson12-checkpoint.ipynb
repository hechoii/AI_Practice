```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

my_path = 'D:/python/AI/img/01/01.JPG'
img = Image.open(my_path)
plt.imshow(img)
im_arr = np.array(img)

print(im_arr.shape)

gray_img = img.convert("L")
im_arr2 = np.array(gray_img)
print(im_arr2.shape)
print(im_arr2.dtype)
```

```python
gray_img = img.convert("L")
plt.imshow(gray_img,cmap="gray")
```

```python
# Bin i nh xm theo yu cu:
img_array = np.array(gray_img)

inverted_img = 255 - img_array # o ngc sng ti
square_img = img_array ** 2 # bnh phng gi tr pixel
clamped_img = np.clip(img_array,100,200)

# Hin th hnh nh:
fig, axis =plt.subplots(2,2, figsize=(10,10)) # to ra khung nh c 4  v size l 10x10
fig.subplots_adjust(wspace=0.2, hspace=0.2) # to khong trng gia cc nh

axis[0, 0].imshow(img_array,cmap="gray")
axis[0, 0].set_title("Anh xam ban dau")

axis[0, 1].imshow(inverted_img,cmap="gray")
axis[0, 1].set_title("Hinh anh da bien doi")

axis[1, 0].imshow(square_img,cmap="gray")
axis[1, 0].set_title("Hinh anh da bien doi")

axis[1, 1].imshow(clamped_img,cmap="gray")
axis[1, 1].set_title("Hinh anh da bien doi")
```
