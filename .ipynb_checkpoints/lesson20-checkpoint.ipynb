```python
import cv2
import numpy as np
from imgtools import *
```

```python
# Tạo một ảnh
img = np.zeros((400,400,3), dtype=np.uint8)
```

```python
display4CV("anh den",img)
```

```python
cv2.line(img, (50,50), (350,50), (0,0,255),1)
display4CV("Duong_thang", img)
```

```python
cv2.circle(img, (150,150),10,(255,180,243),-1)
display4CV("Duong_tron", img)
```

```python
cv2.rectangle(img, (50,50), (350,400), (255,20,173),1)
display4CV("HCN", img)
```

```python
# Ghi text:
content = "PNH"
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,content,(10,300),font,0.6,(255,255,255))
display4CV("chu",img)
```
