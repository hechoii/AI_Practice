```python
import numpy as np
import os
import cv2
import pandas as pd
import pickle
```

```python
# Face detection model:
face_detection_model_path = "./moudules/res10_300x300_ssd_iter_140000_fp16.caffemodel"

# M√¥ t·∫£ ki·∫øn tr√∫c:
face_detection_proto_path = "./moudules/deploy.prototxt"

# M√¥ h√¨nh m√¥ t·∫£ ƒë·∫∑c tr∆∞ng khu√¥n m·∫∑t:
face_descriptor_path = "./moudules/openface.nn4.small2.v1.t7"
```

```python
# S·ª≠ d·ª•ng OpenCV DNN ƒë·ªçc m√¥ h√¨nh nh·∫≠n d·∫°ng khu√¥n m·∫∑t:
detector_model = cv2.dnn.readNetFromCaffe(face_detection_proto_path, face_detection_model_path)

# ƒê·ªçc m√¥ h√¨nh m√¥ t·∫£ ƒë·∫∑c tr∆∞ng khu√¥ng m·∫∑t t·ª´ file Torch:
description_model = cv2.dnn.readNetFromTorch(face_descriptor_path)
```

```python
# Hi·ªÉn th·ªã 1 ·∫£nh:
imgMusk = "img/01/musk11.jpg"
img_orin = cv2.imread(imgMusk)
cv2.imshow("musk",img_orin)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# copy ·∫£nh:
img = img_orin.copy()

# L·∫•y chi·ªÅu cao r·ªông c·ªßa ·∫£nh:
h, w = img.shape[:2]

# Chu·∫©n b·ªã d·ªØ li·ªáu ƒë·∫ßu v√†o cho m√¥ h√¨nh nh·∫≠n di·ªán khu√¥n m·∫∑t:
imgBlob = cv2.dnn.blobFromImage(img,1,(300,300),(104,177,123),swapRB = False, crop = False)

# Thi·∫øt l·∫≠p ƒë·∫ßu v√†o cho m√¥ h√¨nh:
detector_model.setInput(imgBlob)

# Nh·∫≠n di·ªán khu√¥n m·∫∑t:
detections = detector_model.forward()

# Ki·ªÉm tra xem c√≥ nh·∫≠n di·ªán ƒë∆∞·ª£c khu√¥n m·∫∑t kh√¥ng:
if(len(detections)>0):
    # Ch·ªçn khu√¥n m·∫∑t c√≥ ƒë·ªô tinn c·∫≠y cao nh·∫•t:
    i = np.argmax(detections[0,0,:,2]) # Gi√° tr·ªã th·ª© 2 trong 7 gi√° tr·ªã cu·ªëi c√πng -> confidence
    confidence = detections[0,0,i,2]

    # Ki·ªÉm tra ƒë·ªô tin c·∫≠y cao h∆°n 0.5
    if(confidence > 0.5):
        box = detections[0,0,i,3:7] * np.array([w,h,w,h])
        (startx, starty, endx, endy) = box.astype("int")

        # Tr√≠ch xu·∫•t v√πng ·∫£nh chauw khu√¥n m·∫∑t:
        region_img = img[starty : endy, startx : endx]
        #cv2.imshow("hehehe",region_img)
        #cv2.waitKey(0)
        #cv2.destroyAllWindows()

        # Chu·∫©n b·ªã d·ªØu li·ªáu ƒë·∫ßu v√†o cho m√¥ h√¨nh tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng:
        faceBlob = cv2.dnn.blobFromImage(region_img,1/255,(96,96),(0,0,0),swapRB = True,crop = True)

        # Thi·∫øt l·∫≠p m√¥ h√¨nh cho ƒë·∫ßu v√†o
        description_model.setInput(faceBlob)

        # Tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng ra:
        vectors = description_model.forward()

        print(vectors)
        
```

```python
# T·∫°o function ƒë·ªÉ √°p d·ª•ng cho nhi·ªÅu ·∫£nh:
def detect_imgs(img_path):
    img_orin = cv2.imread(img_path)
    
    # copy ·∫£nh:
    img = img_orin.copy()
    
    # L·∫•y chi·ªÅu cao r·ªông c·ªßa ·∫£nh:
    h, w = img.shape[:2]
    
    # Chu·∫©n b·ªã d·ªØ li·ªáu ƒë·∫ßu v√†o cho m√¥ h√¨nh nh·∫≠n di·ªán khu√¥n m·∫∑t:
    imgBlob = cv2.dnn.blobFromImage(img,1,(300,300),(104,177,123),swapRB = False, crop = False)
    
    # Thi·∫øt l·∫≠p ƒë·∫ßu v√†o cho m√¥ h√¨nh:
    detector_model.setInput(imgBlob)
    
    # Nh·∫≠n di·ªán khu√¥n m·∫∑t:
    detections = detector_model.forward()
    
    # Ki·ªÉm tra xem c√≥ nh·∫≠n di·ªán ƒë∆∞·ª£c khu√¥n m·∫∑t kh√¥ng:
    if(len(detections)>0):
        # Ch·ªçn khu√¥n m·∫∑t c√≥ ƒë·ªô tinn c·∫≠y cao nh·∫•t:
        i = np.argmax(detections[0,0,:,2]) # Gi√° tr·ªã th·ª© 2 trong 7 gi√° tr·ªã cu·ªëi c√πng -> confidence
        confidence = detections[0,0,i,2]
    
        # Ki·ªÉm tra ƒë·ªô tin c·∫≠y cao h∆°n 0.5
        if(confidence > 0.5):
            box = detections[0,0,i,3:7] * np.array([w,h,w,h])
            (startx, starty, endx, endy) = box.astype("int")
    
            # Tr√≠ch xu·∫•t v√πng ·∫£nh chauw khu√¥n m·∫∑t:
            region_img = img[starty : endy, startx : endx]
            #cv2.imshow("hehehe",region_img)
            #cv2.waitKey(0)
            #cv2.destroyAllWindows()
    
            # Chu·∫©n b·ªã d·ªØu li·ªáu ƒë·∫ßu v√†o cho m√¥ h√¨nh tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng:
            faceBlob = cv2.dnn.blobFromImage(region_img,1/255,(96,96),(0,0,0),swapRB = True,crop = True)
    
            # Thi·∫øt l·∫≠p m√¥ h√¨nh cho ƒë·∫ßu v√†o
            description_model.setInput(faceBlob)
    
            # Tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng ra:
            vectors = description_model.forward()
    
            return vectors
        
```

```python
# √Åp d·ª•ng cho c√°c ·∫£nh:

# Kh·ªüi t·∫°o dictionary:
data = dict(data=[], label=[])
mydir = "./img/01/train"
folders = os.listdir(mydir)

# Duy·ªát qua t·ª´ng th∆∞ m·ª•c:
for folder in folders:
    path = mydir + "/" + folder
    files = os.listdir(path)

    for fileName in files:
        try:
            vector = detect_imgs(path + "/" + fileName)
    
            # Ki·ªÉm tra vi·ªác tr√≠ch xu·∫•t ƒë∆∞·ª£c kh√¥ng:
            if vector is not None:
                # Th√™m vector v√†o dictionary
                data["data"].append(vector)
                data["label"].append(folder)
                print("compelted")
        except:
            # B·ªè qua n·∫øu g·∫∑p l·ªói:
            pass
```

```python
import numpy as np
import pickle

data = pickle.load(open("./data/face_reg/data_face_feature.pickle", "rb"))

X = data['data']
y = data['label']

print("T·ªïng s·ªë ·∫£nh:", len(X))
print("T·ªïng s·ªë nh√£n:", len(y))

# Ki·ªÉm tra k√≠ch th∆∞·ªõc c·ªßa 1 ph·∫ßn t·ª≠
print("Ki·ªÉu c·ªßa 1 ph·∫ßn t·ª≠ X:", type(X[0]))
print("K√≠ch th∆∞·ªõc ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n:", np.array(X[0]).shape)

# In th·ª≠ 5 nh√£n ƒë·∫ßu ti√™n
print("5 nh√£n ƒë·∫ßu ti√™n:", y[:5])
```

```python
X = data["data"]
y = data["label"]

# L√†m ph·∫≥ng c√°c vector t·ª´ (1,128) ‚Üí (128,)
X = np.array([np.array(v).flatten() for v in X])
y = np.array(y)

print("‚úÖ K√≠ch th∆∞·ªõc d·ªØ li·ªáu:", X.shape)
print("üß† Nh√£n:", np.unique(y))
```

```python

# Chia t·∫≠p train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Train m√¥ h√¨nh KNN (k=3 l√† l·ª±a ch·ªçn t·ªët cho d·ªØ li·ªáu nh·ªè)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# D·ª± ƒëo√°n tr√™n t·∫≠p test
y_pred = knn.predict(X_test)

# ƒê√°nh gi√° hi·ªáu su·∫•t
print("\nüéØ ƒê·ªô ch√≠nh x√°c:", round(metrics.accuracy_score(y_test, y_pred), 4))
print("\nüìä B√°o c√°o ph√¢n lo·∫°i:\n", metrics.classification_report(y_test, y_pred))
```

```python
import pickle
import numpy as np

# ‚úÖ Load m√¥ h√¨nh ƒë√£ l∆∞u
knn = pickle.load(open("./data/face_reg/knn_face_model.pkl", "rb"))
print("‚úÖ ƒê√£ load m√¥ h√¨nh KNN!")

# ‚úÖ face_feature l√† ƒë·∫∑c tr∆∞ng khu√¥n m·∫∑t ƒë·∫ßu v√†o (d·∫°ng (1, 128))
# ƒê·∫£m b·∫£o ƒë√∫ng d·∫°ng 2 chi·ªÅu
if face_feature is None:
    print("‚ùå Kh√¥ng ph√°t hi·ªán ƒë∆∞·ª£c khu√¥n m·∫∑t trong ·∫£nh!")
else:
    # ‚úÖ ƒê·∫£m b·∫£o ƒë√∫ng d·∫°ng 2 chi·ªÅu
    if face_feature.ndim == 1:
        face_feature = face_feature.reshape(1, -1)

    # ‚úÖ D·ª± ƒëo√°n ng∆∞·ªùi v√† ƒë·ªô tin c·∫≠y
    y_pred = knn.predict(face_feature)[0]
    proba = knn.predict_proba(face_feature)[0]
    confidence = np.max(proba)

    # B∆∞·ªõc 3: T√≠nh kho·∫£ng c√°ch g·∫ßn nh·∫•t
    distances, indices = knn.kneighbors(face_feature, n_neighbors=1)
    min_dist = distances[0][0]

    # ----> Chuy·ªÉn ƒë·ªïi kho·∫£ng c√°ch th√†nh ƒë·ªô tin c·∫≠y
    # kho·∫£ng c√°ch c√†ng nh·ªè ‚Üí ƒë·ªô tin c·∫≠y c√†ng cao
    # H√†m logistic ƒë·ªÉ chuy·ªÉn ƒë·ªïi kho·∫£ng c√°ch v·ªÅ 0‚Äì1
    confidence_dis = np.exp(-min_dist)  # ƒë∆°n gi·∫£n, d·ªÖ hi·ªÉu

    if confidence_dis >= 0.5 and confidence == 1:
        y_pred = knn.predict(face_feature)

        # ‚úÖ L·∫•y nh√£n c√≥ ƒë·ªô tin c·∫≠y cao nh·∫•t
        label_confidence = knn.classes_[np.argmax(proba)]
        
        # ‚úÖ In k·∫øt qu·∫£
        print(f"üß† K·∫øt qu·∫£ nh·∫≠n di·ªán: {label_confidence}")
        print(f"üîπ ƒê·ªô tin c·∫≠y: {confidence_dis:.2f}")
    else:
        print(f"ü§î Kh√¥ng ch·∫Øc ch·∫Øn (ƒë·ªô tin c·∫≠y: {confidence_dis*100:.2f}%)")
    
    
```

```python
def predict_face_with_confidence(image_path, model_path="./data/face_reg/knn_face_model.pkl", threshold=0.6):
    """
    H√†m d·ª± ƒëo√°n khu√¥n m·∫∑t t·ª´ ·∫£nh b·∫±ng m√¥ h√¨nh KNN v·ªõi ƒë·ªô tin c·∫≠y.
    - image_path: ƒë∆∞·ªùng d·∫´n ·∫£nh c·∫ßn nh·∫≠n di·ªán
    - model_path: ƒë∆∞·ªùng d·∫´n file m√¥ h√¨nh KNN (.pkl)
    - threshold: ng∆∞·ª°ng tin c·∫≠y (0-1), v√≠ d·ª• 0.6 t∆∞∆°ng ƒë∆∞∆°ng 60%
    """

    # B∆∞·ªõc 1: Load m√¥ h√¨nh KNN
    knn = pickle.load(open(model_path, "rb"))

    face_feature = detect_imgs(image_path)

    if len(face_feature) == 0:
        print("‚ö†Ô∏è Kh√¥ng ph√°t hi·ªán khu√¥n m·∫∑t n√†o trong ·∫£nh.")
        return "unknown", 0.0

    # B∆∞·ªõc 3: T√≠nh kho·∫£ng c√°ch g·∫ßn nh·∫•t
    distances, indices = knn.kneighbors(face_feature, n_neighbors=1)
    min_dist = distances[0][0]

    # ----> Chuy·ªÉn ƒë·ªïi kho·∫£ng c√°ch th√†nh ƒë·ªô tin c·∫≠y
    # kho·∫£ng c√°ch c√†ng nh·ªè ‚Üí ƒë·ªô tin c·∫≠y c√†ng cao
    # H√†m logistic ƒë·ªÉ chuy·ªÉn ƒë·ªïi kho·∫£ng c√°ch v·ªÅ 0‚Äì1
    confidence = np.exp(-min_dist)  # ƒë∆°n gi·∫£n, d·ªÖ hi·ªÉu

    # B∆∞·ªõc 4: Ki·ªÉm tra ng∆∞·ª°ng tin c·∫≠y
    if confidence >= threshold:
        y_pred = knn.predict(face_feature)
        print(f"üßç K·∫øt qu·∫£ nh·∫≠n di·ªán: {y_pred[0]} | ƒê·ªô tin c·∫≠y: {confidence*100:.2f}%")
        return y_pred[0], confidence
    else:
        print(f"ü§î Kh√¥ng ch·∫Øc ch·∫Øn (ƒë·ªô tin c·∫≠y: {confidence*100:.2f}%)")
        return "unknown", confidence
```
