```python
from PIL import Image
import matplotlib.pyplot as plt

my_path = 'D:/python/AI/img/01/01.JPG'
img = Image.open(my_path)
```

```python
# Chuyển đổi backend
plt.switch_backend("tkagg")

# Hiển thị ảnh
plt.imshow(img)
plt.title("Click on the image to select points")

#Sử dụng ginput để chọn điểm trên ảnh
points = plt.ginput(5)

print(points)

plt.show()
```

```python
# vẽ lại các điểm trên
plt.close()
plt.imshow(img)
for point in points:
    x,y = point
    plt.plot(x,y,"r*")
plt.show()
```
