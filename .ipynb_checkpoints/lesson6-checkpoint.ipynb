```python
from PIL import Image
from imgtools import *
from IPython.display import display
```

```python
# Đường dẫn ảnh
my_path = 'D:/python/AI/img/01/01.JPG'
path = 'D:/python/AI/img/01/'
```

```python
img = load_image(my_path)
display(img)
```

```python
gray_img = img.convert("L")
```

```python
display(gray_img)
```

```python
gray_img.save(path + "gray_img.PNG")
```

```python
# phân RGB thành 3 tầng màu khác nhau:
red_band, green_band, blue_band = img.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Merger anhr:
merge_img = Image.merge("RGB", (red_band, green_band, blue_band))
display(merge_img)
```
