```python
import cv2
import numpy as np
```

```python
img = cv2.imread("./img/01/face1.PNG")
cv2.imshow("face",img)
cv2.waitKey(0)
cv2.destroyWindow("face")
```

```python
# Tải mô hình đã được huấn luyện trước:
net = cv2.dnn.readNetFromCaffe("./moudules/deploy.prototxt",
                               "./moudules/res10_300x300_ssd_iter_140000_fp16.caffemodel")

# Chuẩn bị dữ liệu đầu vào:
    # img: là ảnh muốn nhận diện khuôn mặt
    # 1.0: Đây là tỉ lệ co dãn cho hình ảnh
    # (300,300): kích thước ảnh yêu cầu cho đầu vào
    # (104,177,123): màu sắc trung bình
    # swapRB: không hoán đổi màu đỏ và xanh
blob = cv2.dnn.blobFromImage(img, 1.0, (300,300), (104,177,123), swapRB= False)

# Dặt dữ liệu đầu vào cho ảnh:
net.setInput(blob)
```

```python
# Chạy qua từng pixel trong mảng để phát hiện khuôn mặt:
faces = net.forward()
print(faces)
```

```python
# Lấy kích thước của ảnh đầu vào:
h = img.shape[0]
w = img.shape[1]
```

```python
# In thông tin:
print(faces.shape) # [0]: số lượng lớp đầu vào; [1]: số lượng lớp đầu ra;
                   # [2]: số hình ảnh khuôn mặt được phát hiện; [3]: mỗi khuôn mặt có 7 "thông tin"
print(faces[0, 0, 0, ]) # 7 thông tin của ảnh:
                        # 0 số thứ tự của ảnh; 1: là khuôn mặt; 0.9999964: độ tin cậy;
                        # 0.08431688 0.03673201 0.27289385  0.21404085: vị trí của ảnh từ đâu đến đâu
```

```python
# Duyệt từng khuôn mặt được phát hiện:
for i in range(0, faces.shape[2]):
    confidence = faces[0,0,i,2]
    
    # kiểm tra nếu độ tin cậy hơn 0.5:
    if(confidence>0.5):
        # Trích xuất toạ độ
        print(faces[0,0,i,3:7])
        startx = int(faces[0,0,i,3]*w)
        starty = int(faces[0,0,i,4]*h)
        endx = int(faces[0,0,i,5]*w)
        endy = int(faces[0,0,i,6]*h)
        print(startx, starty, endx, endy)

        # Vẽ hcn xung quanh mặt
        cv2.rectangle(img, (startx,starty), (endx,endy), (0,255,0), 2)

        # hiển thị độ tin cây:
        text = "Face: {:.2f}%".format(confidence*100)
        cv2.putText(img, text, (startx,starty-10), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,1,(255,255,255))

# hiển thị hình ảnh:
cv2.imshow("face",img)
cv2.waitKey(0)
cv2.destroyWindow("face")
```
