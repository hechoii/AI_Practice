```python
import cv2
from imgtools import *
```

```python
# Khởi tạo Haar Cascade Classifier nhận diện khuôn mặt
face_cascade = cv2.CascadeClassifier("./data/haarcascade_frontalface_default.xml")
eye_glass_cascade = cv2.CascadeClassifier("./data/haarcascade_eye_tree_eyeglasses.xml")
```

```python
# Đọc ảnh
img = cv2.imread("./data/img_400.JPG")
```

```python
# Chuyển sang ảnh xám
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Nhận diện mặt
faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5, minSize=(30,30))
# Nhận diện mắt
eyes = eye_glass_cascade.detectMultiScale(gray_img,scaleFactor=1.1, minNeighbors=2, minSize=(5,5))
```

```python
# Vẽ hộp chứa khuôn mặt:
for (x, y, w, h) in faces:
    cv2.rectangle(img,(x,y), (x+w,y+h), (0,255,0), 2)

# Vẽ hộp chứa mắt:
for (x, y, w, h) in eyes:
    cv2.circle(img,(int(x+ w/2), int(y+ h/2)), 20, (0,255,255), 2)
```
