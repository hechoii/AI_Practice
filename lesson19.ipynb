```python
import cv2
from imgtools import *
```

```python
my_path = 'D:/python/AI/img/01/01.JPG'
img = cv2.imread(my_path)
print(img)
```

```python
cv2.imshow("test",img)
# Chờ 10s
cv2.waitKey(10000)
# Đóng cửa sổ
cv2.destroyWindow("test")
```

```python
# Tách màu
b,g,r = cv2.split(img)
cv2.imshow("test",img)
cv2.imshow("red",r)
cv2.imshow("green",g)
cv2.imshow("blue",b)
cv2.waitKey(0)
# Đóng cửa sổ
cv2.destroyAllWindows()
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Anh xam",gray_img)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.imshow("Ảnh từ bgr thành rgb",img2)

# Chờ 10s
cv2.waitKey(0)
# Đóng cửa sổ
cv2.destroyAllWindows()
```

```python
# Thay đổi thông số của điểm ảnh:
height, width, z = img.shape
quater_h = height/2
quater_w = width/2

green_color =  (0,255,0)

img3 = img.copy() 

# Thay đổi màu góc trái trên 1/4
for y in range(int(quater_h)):
    for x in range(int(quater_w)):
        img3[y,x] = green_color

cv2.imshow("Xoa goc trai tren",img3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
