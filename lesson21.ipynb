```python
import cv2
import time
```

```python
path_video = 'D:/python/AI/img/01/videocat.mp4'
```

```python
my_vid = cv2.VideoCapture(path_video)
# Tạo cửa sổ để hiện thị
cv2.namedWindow("Video Player", cv2.WINDOW_NORMAL)
```

```python
# Ghi text:
font = cv2.FONT_HERSHEY_SIMPLEX
font_color = (255,255,255)
font_scale = 6
font_thickness = 2
```

```python
# Hiển thị từng khung ảnh 1:
while True:
    #Thời gian trc khi đọc
    start_time = time.time()
    
    ref, frame = my_vid.read()
    # Tháot khi không thể đọc frame
    if not ref:
        break

    # Lấy thời gian sau khi đọc xong
    end_time = time.time()

    fps = 1 / (end_time - start_time)
    
    # Ghi số lượng FPS
    cv2.putText(frame,f"FPS: {fps:.2f}",(10,200),font,font_scale,font_color,font_thickness)
    # Hiển thị khung ảnh
    cv2.imshow("Video Player",frame)

    if(cv2.waitKey(10)==ord("q")):
        break


# Huỷ bỏ player
my_vid.release()
cv2.destroyAllWindows()
```
