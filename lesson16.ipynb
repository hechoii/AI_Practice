```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
my_path = 'D:/python/AI/img/01/01.JPG'
img = Image.open(my_path)
```

```python
gray_img = img.convert("L")
arr_img = np.array(gray_img)
```

```python
# Lm m nh vi b lc gauss:
sigma = 0,5
blurred1_img = ndimage.gaussian_filter(arr_img,sigma,mode="constant")
blurred2_img = ndimage.gaussian_filter(arr_img,1,mode="constant")
blurred3_img = ndimage.gaussian_filter(arr_img,2.8,mode="constant")

# Hin th nh
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(gray_img,cmap="gray")
plt.title("Anh goc")

plt.subplot(2,2,2)
plt.imshow(blurred1_img,cmap="gray")
plt.title("Anh mo voi sigma 0.5")

plt.subplot(2,2,3)
plt.imshow(blurred2_img,cmap="gray")
plt.title("Anh mo voi sigma 1")

plt.subplot(2,2,4)
plt.imshow(blurred3_img,cmap="gray")
plt.title("Anh mo voi sigma 2.8")
```
