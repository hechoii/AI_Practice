```python
from PIL import Image
```

```python
# Thư mục chứa hình ảnh
dir = 'D:/python/AI/img/01'
```

```python
image_path = dir + '/01.JPG'
print(image_path)
```

```python
# Dùng Image.open(path) để đọc ảnh
img01 = Image.open(image_path)
```

```python
# Hiển thị hình ảnh
img01.show()
```

```python
print("Định dạng ảnh ", img01.format)
```

```python
print("Kích thước hình ảnh ", img01.size)
```

```python
# Lưu ảnh
new_image_JPG = dir + '/new_01.JPG'
new_image_PNG = dir + '/new_01.PNG'
img01.save(new_image_PNG)
```

```python
# Đóng tập tin
img01.close()
```
