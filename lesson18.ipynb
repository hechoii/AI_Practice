```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
my_path = 'D:/python/AI/img/01/animals.JPG'
img = Image.open(my_path)
```

```python
gray_img = img.convert("L")
arr_img = np.array(gray_img)
print(arr_img)
```

```python
# Chuyn nh nh phn:
binary_img = 1*(arr_img<220)

# nh nhn
labeled_img, num_features = ndimage.label(binary_img)

# To li subplot vi 3 dng v 1 ct
fig, axs = plt.subplots(3,1, figsize = (50,50))

plt.subplot(3,1,1)
plt.imshow(gray_img, cmap="gray")
plt.title("Anh xam")

plt.subplot(3,1,2)
plt.imshow(binary_img, cmap="gray")
plt.title("Anh nhi phan")

plt.subplot(3,1,3)
plt.imshow(labeled_img, cmap="gray")
plt.title(f"Da danh nhan {num_features} doi tuong")
```

```python
# Chuyn nh nh phn:
binary_img = 1*(arr_img<200)

# Thc hin opening v ni cc i tng vi nhau:
open_img = ndimage.binary_opening(binary_img,structure=np.ones((3,3)))

# nh nhn
labeled_img1, num_features1 = ndimage.label(binary_img)
labeled_img, num_features = ndimage.label(open_img)

# To li subplot vi 3 dng v 1 ct
fig, axs = plt.subplots(4,1, figsize = (50,50))

plt.subplot(4,1,1)
plt.imshow(gray_img, cmap="gray")
plt.title("Anh xam")

plt.subplot(4,1,2)
plt.imshow(binary_img, cmap="gray")
plt.title("Anh nhi phan")

plt.subplot(4,1,3)
plt.imshow(labeled_img1, cmap="gray")
plt.title(f"Da danh nhan {num_features1} doi tuong")

plt.subplot(4,1,4)
plt.imshow(labeled_img, cmap="gray")
plt.title(f"Da danh nhan {num_features} doi tuong sau khi opening")
```
