```python
arr = np.array([[1,9,0],
               [3,8,7],
               [4,1,9]])
sep_theo_hang = np.sort(arr,axis=1) #tang dan,axis=1 hang
print(sep_theo_hang)

sep_theo_cot = -np.sort(-arr,axis=0) #giam dan,axis=0 cot
print(sep_theo_cot)

print(arr.ndim) # so chieu
print(arr.size) # kich thuoc(so phan tu)
print(arr.shape) # ma tran axb
```
